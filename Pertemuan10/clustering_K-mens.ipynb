```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = [
    "cinta romantis mas kawin",
    "kawin lari kawula muda",
    "lomba lari hadiah cincin emas",
    "pencurian uang adalah penyakit masyarakat",
    "penyakit kekurangan uang"
]
```

```python
stopwords_id = [
    "dan", "yang", "di", "ke", "dari", "adalah",
    "itu", "ini", "pada", "untuk", "dengan"
]

vectorizer = TfidfVectorizer(stop_words=stopwords_id)
X = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()

```

```python
k = 2
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_
```

```python
for i, doc in enumerate(documents):
    print(f"Dokumen {i+1} masuk ke Cluster {labels[i]}")
```

```python
top_n = 5
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]

for i in range(k):
    print(f"\nCluster {i} kata dominan:")
    for ind in order_centroids[i, :top_n]:
        print(feature_names[ind])

```
